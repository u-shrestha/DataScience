**operations on groups**

```python
df = pd.read_csv("datasets/artwork_data.csv", nrows=50, index_col= 'id')
df
```

```python
artist_group = df.groupby('artist')

for name, group_df in artist_group:
    print(name)
    print(group_df)
    break
    
```

```python
for name, group_df in artist_group:
    min_yr = group_df['acquisitionYear'].min()
    print("{}: {}.".format(name, min_yr))
```
