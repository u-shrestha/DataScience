### stacking with area and bar plots

```python
sales = pd.read_csv('datasets/sales2.csv', parse_dates = ['Date'])
sales.head()
```

```python
sales['Month'] = sales['Date'].dt.month
sales.head()
```

```python
grp = sales.groupby(['Month', 'Product'])['Amount'].sum()
grp
```

```python
grp = grp.unstack()
grp
```

```python
#total revenue per product
revenue = grp.fillna(0).cumsum()
revenue
```

```python
fig, ax = plt.subplots()
revenue.plot.area(title= 'Revenue 2017 (EUR)', ax= ax)
fig.tight_layout()
```

```python
fig, ax = plt.subplots()

total = sales.groupby('Month')['Amount'].sum().cumsum()
total.plot(ax = ax, style= 'r--', legend= True, label= 'All')

revenue.plot.area(title= 'Revenue 2017 (EUR)',
                  ax= ax, 
                  stacked= False,
                  grid= True)
fig.tight_layout()
```

### top 10 countries to win olympics

```python
ath = pd.read_csv('datasets/athletes.csv')
top = ath.groupby('nationality')[['gold', 'silver', 'bronze']]\
    .sum().sort_values('gold', ascending= False).head(10)
top
```

```python
fig, ax = plt.subplots()
top.plot.bar(color= ['gold', 'silver', 'brown'], ax=ax, stacked= True)
```

### styling a plot

```python
styles= {'a': ':s', 'b': '--v', 'c': '-.o'}

with plt.style.context('seaborn-dark'):
    fig, ax = plt.subplots(figsize= (6,3))
    for columns in styles:
        grp[columns].dropna().plot(ax= ax, style= styles)
        
fig.suptitle('Sales 2017', fontweight= 'bold')
ax.set_title('monthly sales for products a, b and c',
             fontstyle= 'italic',
             fontsize= '10')
ax.set_ylabel('Sales (EUR)')
ax.grid(True)
fig.tight_layout()
```

### scatter plots

```python
f = ath[ath['sex'] == 'female']
m = ath[ath['sex'] == 'male']

tennis_f = f[f['sport'] == 'tennis']
tennis_m = m[m['sport'] == 'tennis']

gymnast_f = f[f['sport'] == 'gymnastics']
gymnast_m = m[m['sport'] == 'gymnastics']
```

```python
gymnast_f.plot.scatter(x= 'weight', y= 'height')
```

```python
fig, ax = plt.subplots()
plots = [(tennis_f, 'red', 'Tennis, women'),
         (tennis_m, 'blue', 'Tennis, men'),
         (gymnast_f, 'orange', 'Gymnastics, women'),
         (gymnast_m, 'green', 'Gymnastics, men')]
for df, color, label in plots:
    df.plot.scatter(ax=ax, 
                    x= 'weight',
                    y= 'height',
                    alpha= .25,
                    color= color,
                    label= label)
    
ax.set_xlabel('weight (kg)')
ax.set_ylabel('height (m)')
ax.set_title('heights and weights of olymics athletes')
```

```python
fig, ax = plt.subplots()
gymnast_m.plot.scatter(ax= ax,
                       x= 'weight',
                       y= 'height',
                       c= 'gold',
                       colormap= 'plasma')
fig.tight_layout()
```

### axes: ticks, ticks label, ranges, legends

```python
styles= {'a': ':s', 'b': '--v', 'c': '-.o'}

with plt.style.context('ggplot'):
    fig, ax = plt.subplots(figsize= (6,3))
    for col in styles:
        grp[col].dropna().plot(ax= ax, style= styles)
        
fig.suptitle('Sales 2017', fontweight= 'bold')
ax.set_title('monthly sales for products a, b and c',
             fontstyle= 'italic',
             fontsize= '10')
ax.set_ylabel('Sales (EUR)')
ax.grid(True)

#x-axis customization

ax.set_xticks(range(1,13))

import calendar

ax.set_xticklabels([calendar.month_name[x] for x in grp.index],
                  rotation= 45,
                  ha= 'right')


fig.tight_layout()
```

```python
styles= {'a': ':s', 'b': '--v', 'c': '-.o'}

with plt.style.context('ggplot'):
    fig, ax = plt.subplots(figsize= (6,3))
    for col in styles:
        grp[col].dropna().plot(ax= ax, style= styles,
                               ylim= (0,20000), legend= True,
                               label= 'Product' + col)
        
fig.suptitle('Sales 2017', fontweight= 'bold')
ax.set_title('monthly sales for products a, b and c',
             fontstyle= 'italic',
             fontsize= '10')
ax.set_ylabel('Sales (EUR)')
ax.grid(True)

#x-axis customization

ax.set_xticks(range(1,13))

import calendar

ax.set_xticklabels([calendar.month_name[x] for x in grp.index],
                  rotation= 45,
                  ha= 'right')

#y-axis customization

def make_yticklabel(tick_value, pos):
    return "{}K".format(tick_value/1000)

from matplotlib.ticker import FuncFormatter
ax.yaxis.set_major_formatter(FuncFormatter(make_yticklabel))

#legend customization

ax.legend(labels= ['Apple', 'Pears', 'Oranges'],
          loc= 'upper left',
          markerfirst= False)

fig.tight_layout()
```
