```python
import json
```

```python
keys_to_use = ['id', 'totalWorks', 'gender', 'mda', 'fc']

def get_records(file_path, keys_to_use):
    
    with open(file_path) as artist_file:
        content = json.load(artist_file)
    record = []
    for field in keys_to_use:
        record.append(content[field])
    return tuple(record)
```

```python
sample_record = get_records("datasets/artists/a/afro-632.json", keys_to_use)
sample_record
```

```python
pd.DataFrame.from_records(sample_record, columns=keys_to_use, index= 'id')
```
