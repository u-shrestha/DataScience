```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

### water dataset

```python
water =  pd.read_csv('datasets/water.csv')
water.head()
```

```python
water.info()
```

**plotting whole dataset in a single plot**

```python
water.plot()
#x-axis - index
```

```python
ax = water.plot() 
```

**plotting dataset columns in diff plots**

```python
ax = water.plot(subplots= True)
```

```python
ax = water.plot(subplots= True, title= 'Water Measurements of the Dutch Coast')
```

**plotting a single column**

```python
ax = water['temp'].plot(title= 'Water Temp')
```

```python
water['datetime'] = pd.to_datetime(water['datetime'])
```

**plotting two columns**

```python
ax = water.plot(title= 'Water Temp', x= 'datetime', y='temp')
```

```python

water = water.set_index(water['datetime'])
water.drop(columns = 'datetime', inplace= True)
water.head()
```

```python
water.index.name = 'Date/Time'
water.rename(columns= {'height': 'Height(cm)'}, inplace= True)

ax = water['Height(cm)'].plot(title= 'tides on the dutch coast', legend= True)
```

**plotting frequency distribution**

```python
water['temp'].plot.hist()
```

```python
water['temp'].plot.box()
```

### nobel dataset

```python
nobel = pd.read_csv('datasets/nobel.csv')
nobel.head()
```

**nobel prize winners by country**

```python
country = nobel['Country'].value_counts()
country
```

```python
top_country = country[:10]
ax = top_country.plot.bar(title= 'top nobel prize winning countries')
```
