### comparing asia and europe's gdp per capita in 2007

### comparing GDP Per Capita growth in the U.S. and China

```python
df['country'].unique()
```

```python
us = df[df['country'] == 'United States']
china = df[df['country'] == 'China']
```

```python
plt.plot(us['year'], us['gdpPerCapita'])
plt.ylabel('GDP per capita')
plt.xlabel('year')
plt.title('GDP Per Capita Growth in the U.S.')
plt.show()
```

```python
plt.plot(china['year'], china['gdpPerCapita'])
plt.ylabel('GDP per capita')
plt.xlabel('year')
plt.title('GDP Per Capita Growth in the China.')
plt.show()
```

```python
plt.plot(china['year'], china['gdpPerCapita'])
plt.plot(us['year'], us['gdpPerCapita'])
plt.ylabel('GDP per capita')
plt.xlabel('year') 
plt.title('GDP Per Capita Growth in the China vs. U.S. (Absolute)')
plt.legend(['China', 'U.S.'])
plt.show()
```

```python
us['gdpPerCapita'].iloc[0]
```

```python
us_growth = us['gdpPerCapita']/us['gdpPerCapita'].iloc[0] * 100
china_growth = china['gdpPerCapita']/china['gdpPerCapita'].iloc[0] * 100
```

```python
plt.plot(china['year'], china_growth)
plt.plot(us['year'], us_growth)
plt.ylabel('GDP per capita growth')
plt.xlabel('year') 
plt.title('GDP Per Capita Growth in the China vs. U.S. (Relative)')
plt.legend(['China', 'U.S.'])
plt.show()
```

### compare the population of the 10 most populous countries in 2007

```python
top10 = data_2007.sort_values('population', ascending= False).head(10)
```

```python
x = range(10)
plt.bar(x, top10['population']/ 10**6)
plt.xticks(x, top10['country'], rotation= 'vertical')
plt.xlabel('countries')
plt.ylabel('population in millions')
plt.title('Top 10 populous countries in 2007')
plt.show()
```

### comparing the gdp of the 10 most populous countries in 2007

```python
gdp_2007 = (top10['gdpPerCapita'] * top10['population']) / 10**9
```

```python
x = range(10)
plt.bar(x, gdp_2007)
plt.xticks(x, top10['country'], rotation= 'vertical')
plt.xlabel('countries')
plt.ylabel('gdp in billions')
plt.title('GDP of top 10 populous countries in 2007')
plt.show()
```

```python
x = np.arange(10)
fig, ax1 = plt.subplots()

width = 0.3 # This is the width of each bar in the bar plot.
plt.xticks(x, top10['country'], rotation='vertical')
population = ax1.bar(x, top10['population']/ 10**6, width)
plt.ylabel('Population')

# ax1.twinx() gives us the same x-axis with the y-axis on the right.
ax2 = ax1.twinx()
gdp = ax2.bar(x + width, gdp_2007, width, color='orange')
plt.ylabel('GDP')
plt.legend([population, gdp],
           ['Population in Millions', 'GDP in Billions'])
figure = plt.gcf() # get current figure
plt.show()
```

### data aggregation

```python
odf = pd.read_csv('datasets/obama.csv', parse_dates= ['year_month'])
odf.head()
```

```python
plt.plot(odf['year_month'], odf['approve_percent'], 'o',
         markersize= 2,
         alpha= 0.4)
plt.show()
```

```python
odf_mean = odf.groupby('year_month').mean()
plt.plot(odf_mean.index, odf_mean['approve_percent'], 'red')
plt.show()
```

```python
plt.plot(odf['year_month'], odf['approve_percent'], 'o',
         markersize= 2,
         alpha= 0.3)
plt.plot(odf_mean.index, odf_mean['approve_percent'], 'red')
plt.show()
```

```python
odf_median = odf.groupby('year_month').median()

plt.plot(odf_mean.index, odf_mean['approve_percent'], 'red')
plt.plot(odf_median.index, odf_median['approve_percent'], 'green')
plt.plot(odf['year_month'], odf['approve_percent'], 'o',
         markersize= 2,
         alpha= 0.3)

plt.legend(['mean', 'median'])
plt.show()
```

### random sampling for  too big dataset

```python
obama = pd.read_csv('datasets/obama_too_big.csv', parse_dates = ['year_month'])
obama.head()
```

```python
obama.shape
```

```python
plt.plot(obama['year_month'], obama['approve_percent'], 'o',
         markersize= 2,
         alpha= 0.3)
plt.show()
```

```python
sample = obama.sample(frac= 0.1)

plt.plot(sample['year_month'], sample['approve_percent'], 'o',
         markersize= 2,
         alpha= 0.3)
plt.show()
```

```python
sample = obama.sample(frac= 0.05)

plt.plot(sample['year_month'], sample['approve_percent'], 'o',
         markersize= 2,
         alpha= 0.3)
plt.show()
```
