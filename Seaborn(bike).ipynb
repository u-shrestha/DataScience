```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bike_df = pd.read_csv('datasets/day.csv')
bike_df.head()
```

```python
bike_df.describe()
```

```python
fig, ax = plt.subplots(figsize= (15,5))
sns.distplot(bike_df['cnt'], bins= 75, kde= False, rug= True);
```

```python
sns.jointplot(x= "instant", y= "cnt",
              data= bike_df, height= 8)
```

```python
sns.jointplot(x= "mnth", y= "cnt",
              data= bike_df, height= 8)
```

```python
sns.pairplot(bike_df, height= 5,
             aspect= 0.9,
             x_vars= ["holiday", "weekday", "workingday"],
             y_vars= ["cnt"],
             kind= "reg",
             hue= "season")
```

```python
sns.jointplot(x= "season", y= "cnt", data= bike_df, height= 8)
```

```python
sns.pairplot(bike_df, height= 5,
             aspect= 0.9,
             x_vars= ["temp", "atemp", "hum", "windspeed"],
             y_vars= ["cnt"],
             hue= "season")
```
