```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
stock_df = pd.read_csv('datasets/stocks.csv')
stock_df.head()
```

```python
stock_df['Date'] = pd.to_datetime(stock_df['Date'])
stock_df.head()
```

### comparing stock prices

```python
fig = plt.figure(figsize= (10,6))

ax1 = fig.add_axes([0,0,1,1])
ax1.plot(stock_df['Date'], stock_df['AAPL'],
        color = 'green')
ax1.set_title('apple vs. ibm', fontsize= 25)

ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])
ax2.plot(stock_df['Date'], stock_df['IBM'],
        color = 'blue')
```

```python
fig = plt.figure(figsize= (10,6))
fig.suptitle('stock price comparison 2007 to 2017', fontsize= 20)

ax1 = fig.add_subplot(221)
ax1.plot(stock_df['Date'], stock_df['MSFT'],
        color = 'green')
ax1.set_title('Microsoft')

ax2 = fig.add_subplot(222)
ax2.plot(stock_df['Date'], stock_df['GOOG'],
        color = 'purple')
ax2.set_title('Google')

ax3 = fig.add_subplot(223)
ax3.plot(stock_df['Date'], stock_df['SBUX'],
        color = 'magenta')
ax3.set_title('Starbucks')

ax4 = fig.add_subplot(224)
ax4.plot(stock_df['Date'], stock_df['CVX'],
        color = 'orange')
ax4.set_title('Chevron')

# plt.tight_layout()
```

### drawing shapes with the patches module

```python
import matplotlib.patches as patches
```

```python
fig, ax = plt.subplots()

ax.add_patch(
    patches.Rectangle(
    (0.1, 0.1),
    0.5,
    0.5,
    fill= False))
plt.show()
```

```python
fig, ax = plt.subplots()

ax.add_patch(
    patches.Rectangle(
    (0.1, 0.1),
    0.5,
    0.5,
    facecolor= 'grey',
    edgecolor= 'orange'))
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect= 'equal')

for p in [
    patches.Rectangle((0.1, 0.1),
                      0.3, 0.6,
                      hatch = '\\',
                      fill= False),
    patches.Rectangle((0.5,0.1),
                     0.3, 0.6,
                     hatch= '.')
]:
    ax.add_patch(p)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect= 'equal')

for p in [
    patches.Rectangle((0.1, 0.1),
                      0.2, 0.6,
                      linestyle= 'solid',
                      fill= False),
     patches.Rectangle((0.4, 0.1),
                      0.2, 0.6,
                      linestyle= 'dashed',
                      fill= False),
     patches.Rectangle((0.7, 0.1),
                      0.2, 0.6,
                      linestyle= 'dashdot',
                      fill= False),
     patches.Rectangle((1.0, 0.1),
                      0.2, 0.6,
                      linestyle= 'dotted',
                      fill= False),
    
]:
    ax.add_patch(p)
    
ax.set_xlim(0,1.5)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect= 'equal')

for p in [
    patches.Circle((0.2, 0.4),
                      0.1,
                      hatch = '\\',
                      fill= False),
    patches.Circle((0.7,0.4),
                     0.2,
                     hatch= '.',
                    facecolor= 'aqua')
]:
    ax.add_patch(p)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect= 'equal')

polygon = patches.Polygon([[0.1,0.1],
                         [0.2,0.8],
                         [0.5,0.8],
                         [0.8,0.2],
                         [0.5,0.3]],
                         fill= False)

ax.add_patch(polygon)
plt.show()
```

### annotations

```python
#representing the same data in different units

def fahrenheit2celsius(f):
    return (f-32)*5/9
```

```python
fig, ax_tempF = plt.subplots()

fig.set_figwidth(12)
fig.set_figheight(6)

ax_tempF.set_xlabel('Date')
ax_tempF.tick_params(axis = 'x',
                     bottom = False,
                     labelbottom = False)

ax_tempF.set_ylabel('Temp (F)',
                   color= 'red',
                   size= 'x-large')
ax_tempF.tick_params(axis= 'y',
                    labelcolor= 'red',
                    labelsize= 'large')

ax_tempF.plot(weather['Date'],
              weather['TempAvgF'],
              color= 'green')

ax_tempC = ax_tempF.twinx() 

ymin, ymax = ax_tempF.get_ylim()

ax_tempC.set_ylim(fahrenheit2celsius(ymin),
                  fahrenheit2celsius(ymax))

ax_tempC.set_ylabel('temp (C)',
                   color= 'blue',
                   size= 'x-large')
ax_tempC.tick_params(axis= 'y',
                    labelcolor= 'blue',
                    labelsize= 'large')
plt.show()
```
