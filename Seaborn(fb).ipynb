```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### building trellis plots

**FacetGrids is used to visualize relationships between multiple variables separately**

```python
fb_df = pd.read_csv('datasets/dataset_Facebook.csv',
                    sep= r'\s*;\s*', engine= 'python')
fb_df.head()
```

```python
fb_df.info()
```

```python
fb_df.describe()
```

```python
g = sns.FacetGrid(fb_df, col= 'Type')
g.map(plt.hist, 'Total Interactions')
```

```python
g = sns.FacetGrid(fb_df, col= 'Type')
g.map(plt.hist, 'Total Interactions')

plt.xlim(-20, 3000)
plt.ylim(0, 30)
```

```python
g = sns.FacetGrid(fb_df, col= 'Type')
g.map(plt.scatter, 'Lifetime Post Total Reach', 'Lifetime Engaged Users')
```

```python
g = sns.FacetGrid(fb_df, col= 'Type', hue= 'Paid')
g.map(plt.scatter, 'Lifetime Post Total Reach', 'Lifetime Engaged Users', alpha= 0.7)
g.add_legend()
```

```python
g = sns.FacetGrid(fb_df, row= 'Category',
                  col= 'Type', hue= 'Paid', margin_titles= True)
g.map(sns.regplot, 'Post Month', 'Total Interactions', fit_reg= False)
g.add_legend()
plt.ylim(0, 2500)
```

```python
g = sns.FacetGrid(fb_df, row= 'Category',
                  col= 'Type', hue= 'Paid',
                  margin_titles= True, 
                  gridspec_kws= {'width_ratios': [5,3,2,2]})
g.map(sns.regplot, 'Post Month', 'Total Interactions', fit_reg= False)
g.add_legend()
plt.ylim(0, 2500)
```

```python
h= {0: 'red', 1: 'green'}
g = sns.FacetGrid(fb_df, row= 'Category',
                  col= 'Type', hue= 'Paid',
                  margin_titles= True, palette= h)
g.map(sns.regplot, 'Post Month', 'Total Interactions', fit_reg= False)
g.add_legend()
plt.ylim(0, 2500)
```

```python
h= {0: 'red', 1: 'green'}
i= {'marker': ["^", "v"]}

g = sns.FacetGrid(fb_df, row= 'Category',
                  col= 'Type', hue= 'Paid',
                  margin_titles= True, palette= h,
                  hue_kws= i)
g.map(sns.regplot, 'Post Month', 'Total Interactions', fit_reg= False)
g.add_legend()
plt.ylim(0, 2500)
```

```python
g= sns.FacetGrid(fb_df, col= 'Type')
g.map(sns.barplot, "Post Weekday", "Page total likes")
```

```python
g= sns.FacetGrid(fb_df, col= 'Type', 
                col_order= ["Video", "Photo", "Status", "Link"])
g.map(sns.barplot, "Post Weekday", "Page total likes")
```

```python
g= sns.FacetGrid(fb_df, col= 'Post Month', col_wrap= 3)
g.map(sns.barplot, "Post Weekday", "Page total likes")
```

### customizing facet grids

```python
g= sns.FacetGrid(fb_df, col= 'Post Month', col_wrap= 3)
g.map(sns.barplot, "Post Weekday", "Page total likes",
      color= "#334488", edgecolor= "red", lw= .5)
```

```python
g= sns.FacetGrid(fb_df, col= 'Post Month', col_wrap= 3)
g.map(sns.barplot, "Post Weekday", "Page total likes",
      color= "#aaff33", edgecolor= "white", lw= .5)
g.fig.subplots_adjust(wspace= 1, hspace= 0.5);
```

```python
g= sns.FacetGrid(fb_df, col= 'Post Month', col_wrap= 3)
g.map(sns.barplot, "Post Weekday", "Page total likes",
      color= "#eedff0", edgecolor= "white", lw= .5)

g.set_axis_labels("Post every week", "Total likes");
```

```python
g= sns.FacetGrid(fb_df, col= 'Post Month', col_wrap= 3)
g.map(sns.barplot, "Post Weekday", "Page total likes",
      color= "#13f0ff", edgecolor= "white", lw= .5)
g.set(yticks= [40000, 80000, 120000]);
```

```python
g= sns.FacetGrid(fb_df, col= 'Post Month', col_wrap= 3)
g.map(sns.barplot, "Post Weekday", "Page total likes",
      color= "#13f0ff", edgecolor= "white", lw= .5)

g.set_axis_labels("Post every week", "Total likes");

g.set(yticks= [40000, 80000, 120000]);
g.set(xlim= (0,6), ylim= (0, 80000));
```

### pairgrids

```python
v = sns.PairGrid(fb_df,
                 vars= ["comment", "like", "share"])
v.map(plt.scatter)
```

```python
v = sns.PairGrid(fb_df,
                 vars= ["comment", "like", "share"])
v.map_offdiag(plt.scatter)
v.map_diag(sns.kdeplot)
```

```python
v = sns.PairGrid(fb_df,
                 vars= ["comment", "like", "share"],
                 hue= "Type")
v.map_offdiag(plt.scatter)
v.map_diag(sns.kdeplot)
v.add_legend()
```

```python
v = sns.PairGrid(fb_df,
                 vars= ["comment", "like", "share"])
v.map_upper(plt.scatter)
v.map_diag(sns.kdeplot)
v.map_lower(sns.regplot)
```

```python
v = sns.PairGrid(fb_df,
                 x_vars= ["comment", "like", "share"],
                 y_vars = ["Post Month", "Post Weekday", "Post Hour"],
                 hue= "Type")
v.map(plt.scatter)
v.add_legend()
```
