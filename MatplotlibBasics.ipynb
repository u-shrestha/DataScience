```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### basics

```python
x = np.linspace(start= 0, stop= 10, num= 50)

plt.plot(x, np.sin(x))
plt.xlabel('x', fontsize= 15, color= 'green')
plt.ylabel('sin(x)', fontsize= 15, color= 'green')
```

```python
plt.tick_params(axis= 'y',
                color= 'red',
                labelcolor= 'blue',
                labelsize= 'xx-large')
```

```python
plt.tick_params(axis= 'x',
                bottom= False,
                labelbottom= False)
```

```python
plt.plot(x, np.sin(x), label= 'sine curve')
plt.legend()
```

```python
plt.plot(x, np.sin(x), label= 'sine curve')
plt.legend()
plt.title('the sine curve')
plt.xlim(1,5)
```

```python
plt.plot(x, np.sin(x), label= 'sine curve')
plt.legend()
plt.title('the sine curve')
plt.xlim(1,5)
plt.ylim(-1, 0.5)
```

```python
#flipped vertically

plt.plot(x, np.sin(x), label= 'sine curve')
plt.legend()
plt.title('the sine curve')
plt.ylim(-1, 0.5)
plt.xlim(5,1)
```

```python
#flipped horizontally

plt.plot(x, np.sin(x), label= 'sine curve')
plt.legend()
plt.title('the sine curve')
plt.ylim(0.5, -1)
plt.xlim(5,1)
```

```python
plt.plot(x, np.sin(x), label= 'sine curve')
plt.plot(x, np.cos(x), label= 'cosine curve')
plt.legend()
plt.title('sine and cosine curves')
```

```python
plt.plot(x, np.sin(x), label= 'sine curve', color= 'green')
plt.plot(x, np.cos(x), label= 'cosine curve', color= 'm')
plt.legend(loc= 'upper center', ncol= 2, facecolor= 'grey')
plt.title('sine and cosine curves')
```

```python
x = np.linspace(start= 0, stop= 10, num= 50)

fig = plt.figure()
ax1 = fig.add_axes([0, 0.6, 1, 0.4])
ax1.plot(x, np.sin(x))
ax1.set_xlabel('x', fontsize= 15, color= 'r')
ax1.set_ylabel('sin(x)', fontsize= 15, color= 'r')

ax2 = fig.add_axes([0, 0, 0.8, 0.4])
ax2.plot(x, np.cos(x))
ax2.set_xlabel('x', fontsize= 15, color= 'r')
ax2.set_ylabel('cos(x)', fontsize= 15, color= 'r')

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.5, 0.5, 0.4, 0.4])
plt.show()
```

```python
ax1 = plt.subplot2grid((2,3), (0,0))
ax1.plot(x,np.sin(x))
ax1.set_label('sine curve')

ax2 = plt.subplot2grid((2,3), (0,1))
ax2.plot(x,np.cos(x))
ax2.set_label('cosine curve')

ax3 = plt.subplot2grid((2,3), (0,2), rowspan= 2)
ax3.plot([1, 2, 3, 4],
         [2, 4, 6, 8])
ax3.set_label('straight line')
ax3.yaxis.tick_right()

ax4 = plt.subplot2grid((2,3), (1,0), colspan= 2)
ax4.plot(x, np.exp2(x))
ax4.set_label('exponential curve')

plt.tight_layout()
plt.show()
```

### watermarks

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label= 'sine curve')
ax.plot(x, np.cos(x), label= 'cosine curve')
ax.legend()
ax.set_title('sine and cosine curves')
ax.text(1, 0,
        'do not distribute',
        fontsize= 30,
        color= 'red',
        ha= 'left', #horizontal alignment
        va= 'top',
        alpha= 0.4)
```

```python
fig = plt.figure(figsize= (6,4))

fig.text(0.15, 0.4,
         'do not distribute',
         fontsize= 40,
         color= 'gray',
         ha= 'left',
         va= 'bottom',
         alpha= 0.4)

ax1 = fig.add_subplot(121)
ax1.plot(x,np.sin(x))
ax1.set_label('sine curve')

ax2 = fig.add_subplot(122)
ax2.plot(x,np.cos(x))
ax2.set_label('cosine curve')

plt.tight_layout()
plt.show()
```
