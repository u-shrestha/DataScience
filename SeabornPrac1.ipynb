```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
ath = pd.read_csv('datasets/athletes.csv')
ath = ath[ath['sport'].isin(['gymnastics', 'triathlon', 'badminton', 'wrestling'])]
ath.head()
```

### relational plots

```python
sns.relplot(data= ath, x= 'weight', y= 'height')
```

```python
sns.relplot(data= ath, x= 'weight', y= 'height', col= 'sport')
```

```python
sns.relplot(data= ath, x= 'weight', y= 'height', col= 'sex')
```

```python
sns.relplot(data= ath, x= 'weight', y= 'height',
            col= 'sport', col_wrap= 2, hue= 'sex')
```

**correlation between height and weight among athletes**

```python
sns.relplot(x= 'height', y= 'weight',
          data= ath)
```

**correlation between height and weight among athletes based on sex**

```python
sns.relplot(x= 'height', y= 'weight',
          data= ath, hue= 'sex')
```

**correlation between height and weight among athletes based on sports**

```python
sns.relplot(x= 'height', y= 'weight',
          data= ath, hue= 'sport')
```

```python
country = pd.read_csv('datasets/countries2.csv')
country.head()
```

### categorical plots

```python
sns.catplot(data= ath, x= 'sport', y= 'weight', hue= 'sex')
```

```python
sns.catplot(data= ath, x= 'sport', y= 'weight',
            hue= 'sex', kind= 'boxen')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

**relationship between tip amt and day of the week**

```python
sns.catplot(x= 'day', y= 'tip', data= tips)
```

**swarm plots**

```python
sns.catplot(x= 'day', y= 'tip',
           kind= 'swarm', data= tips)
```

**who gets the higher tips: male or female**

```python
sns.catplot(x= 'day', y= 'tip',
           hue= 'sex', data= tips)
```

```python
sns.catplot(x= 'day', y= 'tip',
           hue= 'sex', data= tips, kind= 'swarm')
```

**boxplot**

```python
sns.catplot(x= 'day', y= 'tip',
           data= tips, kind= 'box')
```

```python
sns.boxplot(x= 'day', y= 'tip',
           data= tips)
```

```python
sns.catplot(x= 'day', y= 'tip',
           hue= 'sex', data= tips, kind= 'box')
```

**violin plots**

```python
sns.catplot(x= 'day', y= 'tip',
           hue= 'sex', data= tips, kind= 'violin')
```

**types of sports and what does participation look like**

```python
sns.countplot(x= ath['sport'], 
              order= ath['sport'].value_counts().index)
```

### distribution of dataset

**what does total bill look like in the restaurant**

```python
sns.distplot(tips['total_bill'], kde= False)
```

**KDE plot**

```python
sns.kdeplot(tips['total_bill'], shade= True)
```

**relationship between total_bill and tip**

**bivariate plot- 2D KDE plot**

```python
sns.jointplot(x= tips['total_bill'], y= tips['tip'], kind= 'kde')
```
