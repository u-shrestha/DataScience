```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
income = pd.read_csv('adult.csv')
income.sample(10)
```

```python
income.shape
```

```python
income = income.replace('?',np.nan)
income.sample(10)
```

```python
income.isna().sum()
```

```python
income = income.dropna()
income.isna().sum()
```

```python
income.shape
```

```python
plt.figure(figsize=(10,7))
pd.value_counts(income['marital-status']).plot.bar()
plt.show()

plt.figure(figsize=(10,7))
pd.value_counts(income['occupation']).plot.bar()
plt.show()

plt.figure(figsize=(10,7))

pd.value_counts(income['gender']).plot.bar()

plt.show()
```

```python
Gender = ['Male', 'Female']
```

```python
from sklearn import preprocessing

label_encoding = preprocessing.LabelEncoder()
label_encoding = label_encoding.fit(Gender)
```

```python
income['gender'] = label_encoding.transform(income['gender'].astype(str))
```

```python
label_encoding.classes_
```

```python
income.head(10)
```

```python
one_hot_encoding = preprocessing.OneHotEncoder()
one_hot_encoding = one_hot_encoding.fit(income['marital-status'].values.reshape(-1,1))
```

```python
one_hot_encoding.categories_
```

```python
# one hot encoder expects the input features to be in 2D and hence reshaped

one_hot_labels = one_hot_encoding.transform(income['marital-status'].values.reshape(-1,1)).toarray()
one_hot_labels
```

```python
labels_df = pd.DataFrame()

labels_df['marital-status_divorced'] = one_hot_labels[:,0]
labels_df['marital-status_married-af-spouse'] = one_hot_labels[:,1]
labels_df['marital-status_married-civ-spouse'] = one_hot_labels[:,2]
labels_df['marital-status_marriage-spouse-absent'] = one_hot_labels[:,3]
labels_df['marital-status_never-married'] = one_hot_labels[:,4]
labels_df['marital-status_separated'] = one_hot_labels[:,5]
labels_df['marital-status_widowed'] = one_hot_labels[:,6]

labels_df.head()
```

```python
#drop original marital-status col and used one hot encoded instead

encoded = pd.concat([income, labels_df], axis=1)
encoded.drop('marital-status', axis=1, inplace = True)
encoded.sample(10)
```

```python
# one hot encoding using pandas

income = pd.get_dummies(encoded, columns = ['occupation'])
income.sample(10)
```
