### autocorrelation

**measures the relationship between a variable's current value and past value**

```python
grand_canyon = pd.read_csv('datasets/grand_canyon_visits.csv')
grand_canyon.head()
```

```python
grand_canyon.shape
```

*when using the data for autocorrelation, large numbers may result in an overflow. To avoid this, divide the values by 1000*

```python
grand_canyon['NumVisits'] = grand_canyon['NumVisits']/1000
grand_canyon.describe()
```

```python
plt.figure(figsize= (16,8))
plt.acorr(grand_canyon['NumVisits'], maxlags= 20)
plt.show()
```

```python
plt.figure(figsize= (16,8))
lags, c, vlines, hline = plt.acorr(grand_canyon['NumVisits'],
                                   maxlags= 20)
print('lags: ', lags)
print('\n c: ', c)
print('\n vlines: ', vlines)
print('\n hlines: ', hline)
```

### stacked plots

```python
x = np_df['Year']

y = np.vstack([np_df['Badlands'],
               np_df['GrandCanyon'],
               np_df['BryceCanyon']])
y
```

```python
labels = ['Badlands', 'GrandCanyon', 'BryceCanyon']

plt.stackplot(x,y,
              labels= labels)
plt.legend(loc= 'upper left')
plt.show()
```

```python
colors = ['sandybrown', 'tomato', 'skyblue']

plt.stackplot(x,y,
              labels= labels,
              colors= colors,
              edgecolor = 'grey')
plt.legend(loc= 'upper left')
plt.show()
```
